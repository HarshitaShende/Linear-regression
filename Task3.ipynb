```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
df = pd.read_csv("D:\project\Elevate\Task3+4\Housing.csv")
```

```python
df.dropna(inplace=True)
```

```python
X_simple = df[['price']]  
y = df['price']   
```

```python
X_multiple = df[['price', 'area']]  
```

```python
X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(X_simple, y, test_size=0.2, random_state=42)
X_train_m, X_test_m, y_train_m, y_test_m = train_test_split(X_multiple, y, test_size=0.2, random_state=42)
```

```python
model_simple = LinearRegression()
model_simple.fit(X_train_s, y_train_s)

model_multiple = LinearRegression()
model_multiple.fit(X_train_m, y_train_m)
```

```python
y_pred_s = model_simple.predict(X_test_s)

mae_s = mean_absolute_error(y_test_s, y_pred_s)
mse_s = mean_squared_error(y_test_s, y_pred_s)
r2_s = r2_score(y_test_s, y_pred_s)

print("\nðŸ“Š Simple Linear Regression Metrics:")
print(f"MAE: {mae_s:.2f}")
print(f"MSE: {mse_s:.2f}")
print(f"RÂ² Score: {r2_s:.2f}")
```

```python
y_pred_m = model_multiple.predict(X_test_m)

mae_m = mean_absolute_error(y_test_m, y_pred_m)
mse_m = mean_squared_error(y_test_m, y_pred_m)
r2_m = r2_score(y_test_m, y_pred_m)

print("\nðŸ“Š Multiple Linear Regression Metrics:")
print(f"MAE: {mae_m:.2f}")
print(f"MSE: {mse_m:.2f}")
print(f"RÂ² Score: {r2_m:.2f}")
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(X_test_s, y_test_s, color='blue', label='Actual')
plt.plot(X_test_s, y_pred_s, color='red', linewidth=2, label='Predicted Line')
plt.xlabel('Feature1')
plt.ylabel('Target')
plt.title('Simple Linear Regression Line')
plt.legend()
plt.show()

# Interpret Coefficients
print("\nðŸ§  Coefficient Interpretation (Simple):")
print(f"Intercept: {model_simple.intercept_}")
print(f"Coefficient: {model_simple.coef_[0]} â€” For each unit increase in feature1, target changes by this amount.")

print("\nðŸ§  Coefficient Interpretation (Multiple):")
print(f"Intercept: {model_multiple.intercept_}")
for feature, coef in zip(X_multiple.columns, model_multiple.coef_):
    print(f"{feature}: {coef:.4f} â€” Impact on target per unit change.")
```
